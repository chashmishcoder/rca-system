## Task 1: Core Ontology Design (OWL)

We'll design a comprehensive ontology that covers both manufacturing (AI4I) and transportation (MetroPT) domains with the following core concepts:

### Core Classes:
- **Equipment**: Manufacturing machines, metro systems
- **Component**: Parts, subsystems
- **Sensor**: Temperature, pressure, current sensors
- **Failure**: Different failure types and modes
- **MaintenanceAction**: Preventive, corrective actions
- **Anomaly**: Statistical and domain-specific anomalies

### Domain Extensions:
- **Manufacturing Domain**: Tools, processes, quality types
- **Transportation Domain**: Stations, routes, operational modes

````python
# =================================================================
# TASK 1: CORE ONTOLOGY DESIGN USING OWL
# =================================================================

# Create the main ontology
onto = get_ontology("http://www.semanticweb.org/predictive_maintenance.owl")

with onto:
    # =============================================
    # CORE CLASSES DEFINITION
    # =============================================
    
    # Base Equipment Class
    class Equipment(Thing):
        """Base class for all equipment in predictive maintenance domain"""
        pass
    
    # Component Class
    class Component(Thing):
        """Parts or subsystems of equipment"""
        pass
    
    # Sensor Class  
    class Sensor(Component):
        """Sensors that monitor equipment state"""
        pass
    
    # Failure Class
    class Failure(Thing):
        """Different types of failures that can occur"""
        pass
    
    # Maintenance Action Class
    class MaintenanceAction(Thing):
        """Actions taken for maintenance"""
        pass
    
    # Anomaly Class
    class Anomaly(Thing):
        """Detected anomalies in system behavior"""
        pass
    
    # Process Class
    class Process(Thing):
        """Operational processes"""
        pass
    
    # =============================================
    # MANUFACTURING DOMAIN EXTENSION (AI4I)
    # =============================================
    
    class ManufacturingEquipment(Equipment):
        """Manufacturing machines and tools"""
        pass
    
    class ManufacturingProcess(Process):
        """Manufacturing operational processes"""
        pass
        
    class Tool(Component):
        """Manufacturing tools"""
        pass
        
    class QualityType(Thing):
        """Product quality classifications"""
        pass
    
    # Manufacturing Failure Types
    class ToolWearFailure(Failure):
        """Tool wear related failures (TWF)"""
        pass
        
    class HeatDissipationFailure(Failure):
        """Heat dissipation failures (HDF)"""
        pass
        
    class PowerFailure(Failure):
        """Power related failures (PWF)"""
        pass
        
    class OverstrainFailure(Failure):
        """Overstrain failures (OSF)"""
        pass
        
    class RandomFailure(Failure):
        """Random failures (RNF)"""
        pass
    
    # =============================================
    # TRANSPORTATION DOMAIN EXTENSION (MetroPT)
    # =============================================
    
    class TransportationEquipment(Equipment):
        """Transportation systems and vehicles"""
        pass
        
    class TransportationProcess(Process):
        """Transportation operational processes"""
        pass
    
    class Station(Thing):
        """Metro stations or stops"""
        pass
        
    class Route(Thing):
        """Transportation routes"""
        pass
        
    class OperationalMode(Thing):
        """Different operational modes of transportation systems"""
        pass
    
    # Transportation Specific Components
    class Compressor(Component):
        """Air compressor systems"""
        pass
        
    class ElectricValve(Component):
        """Electric valve systems"""
        pass
        
    class Tower(Component):
        """Tower systems"""
        pass
    
    # =============================================
    # SENSOR SPECIALIZATIONS
    # =============================================
    
    class TemperatureSensor(Sensor):
        """Temperature monitoring sensors"""
        pass
        
    class PressureSensor(Sensor):
        """Pressure monitoring sensors"""
        pass
        
    class CurrentSensor(Sensor):
        """Electric current sensors"""
        pass
        
    class FlowSensor(Sensor):
        """Flow rate sensors"""
        pass
        
    class PositionSensor(Sensor):
        """GPS and position sensors"""
        pass
    
    # =============================================
    # MAINTENANCE ACTION TYPES
    # =============================================
    
    class PreventiveMaintenance(MaintenanceAction):
        """Scheduled preventive maintenance"""
        pass
        
    class CorrectiveMaintenance(MaintenanceAction):
        """Corrective maintenance after failure"""
        pass
        
    class PredictiveMaintenance(MaintenanceAction):
        """Predictive maintenance based on data"""
        pass
    
    # =============================================
    # ANOMALY TYPES
    # =============================================
    
    class StatisticalAnomaly(Anomaly):
        """Statistically detected anomalies"""
        pass
        
    class ThresholdAnomaly(Anomaly):
        """Threshold-based anomalies"""
        pass
        
    class PatternAnomaly(Anomaly):
        """Pattern-based anomalies"""
        pass

print("âœ… Core ontology classes defined")
print(f"ğŸ“Š Total classes created: {len(list(onto.classes()))}")

# Display the class hierarchy
print("\nğŸ—ï¸  ONTOLOGY CLASS HIERARCHY:")
print("=" * 50)

def print_class_hierarchy(cls, indent=0):
    """Recursively print class hierarchy"""
    print("  " * indent + f"ğŸ“‹ {cls.name}")
    for subclass in cls.subclasses():
        print_class_hierarchy(subclass, indent + 1)

# Print hierarchy starting from Thing
for cls in onto.classes():
    if not cls.ancestors() or Thing in cls.ancestors():
        print_class_hierarchy(cls)
````

````python
# =================================================================
# TASK 1.2: OBJECT PROPERTIES & DATA PROPERTIES DEFINITION
# =================================================================

with onto:
    # =============================================
    # OBJECT PROPERTIES (RELATIONSHIPS)
    # =============================================
    
    # Equipment relationships
    class hasComponent(ObjectProperty):
        """Equipment has components"""
        domain = [Equipment]
        range = [Component]
    
    class isPartOf(ObjectProperty):
        """Component is part of equipment"""
        domain = [Component]
        range = [Equipment]
        inverse_property = hasComponent
    
    class hasSensor(ObjectProperty):
        """Equipment/Component has sensors"""
        domain = [Equipment | Component]
        range = [Sensor]
    
    class monitors(ObjectProperty):
        """Sensor monitors equipment/component"""
        domain = [Sensor]
        range = [Equipment | Component]
        inverse_property = hasSensor
    
    # Failure relationships
    class hasFailure(ObjectProperty):
        """Equipment experiences failure"""
        domain = [Equipment]
        range = [Failure]
    
    class causedBy(ObjectProperty):
        """Failure caused by component/anomaly"""
        domain = [Failure]
        range = [Component | Anomaly]
    
    class requiresMaintenance(ObjectProperty):
        """Failure requires maintenance action"""
        domain = [Failure]
        range = [MaintenanceAction]
    
    # Process relationships
    class runsOn(ObjectProperty):
        """Process runs on equipment"""
        domain = [Process]
        range = [Equipment]
    
    class hasProcess(ObjectProperty):
        """Equipment runs processes"""
        domain = [Equipment]
        range = [Process]
        inverse_property = runsOn
    
    # Anomaly relationships
    class hasAnomaly(ObjectProperty):
        """Equipment/Component has anomaly"""
        domain = [Equipment | Component]
        range = [Anomaly]
    
    class detectedBy(ObjectProperty):
        """Anomaly detected by sensor"""
        domain = [Anomaly]
        range = [Sensor]
    
    class precedes(ObjectProperty):
        """Anomaly precedes failure"""
        domain = [Anomaly]
        range = [Failure]
    
    # =============================================
    # MANUFACTURING SPECIFIC RELATIONSHIPS
    # =============================================
    
    class hasQualityType(ObjectProperty):
        """Manufacturing equipment produces quality type"""
        domain = [ManufacturingEquipment]
        range = [QualityType]
    
    class usesTool(ObjectProperty):
        """Manufacturing process uses tools"""
        domain = [ManufacturingProcess]
        range = [Tool]
    
    # =============================================
    # TRANSPORTATION SPECIFIC RELATIONSHIPS
    # =============================================
    
    class operatesAt(ObjectProperty):
        """Transportation equipment operates at station"""
        domain = [TransportationEquipment]
        range = [Station]
    
    class followsRoute(ObjectProperty):
        """Transportation equipment follows route"""
        domain = [TransportationEquipment]
        range = [Route]
    
    class hasOperationalMode(ObjectProperty):
        """Transportation equipment has operational mode"""
        domain = [TransportationEquipment]
        range = [OperationalMode]
    
    # =============================================
    # DATA PROPERTIES (ATTRIBUTES)
    # =============================================
    
    # Identifiers
    class hasID(DataProperty):
        """Unique identifier"""
        domain = [Thing]
        range = [str]
    
    class hasName(DataProperty):
        """Name or label"""
        domain = [Thing]
        range = [str]
    
    # Temporal properties
    class hasTimestamp(DataProperty):
        """Timestamp of occurrence"""
        domain = [Thing]
        range = [str]  # Store as ISO format string
    
    class hasStartTime(DataProperty):
        """Start time"""
        domain = [Process | MaintenanceAction]
        range = [str]  # Store as ISO format string
    
    class hasEndTime(DataProperty):
        """End time"""
        domain = [Process | MaintenanceAction]
        range = [str]  # Store as ISO format string
    
    # Sensor readings
    class hasValue(DataProperty):
        """Sensor reading value"""
        domain = [Sensor]
        range = [float]
    
    class hasUnit(DataProperty):
        """Unit of measurement"""
        domain = [Sensor]
        range = [str]
    
    # Temperature specific
    class hasTemperature(DataProperty):
        """Temperature value"""
        domain = [TemperatureSensor]
        range = [float]
    
    # Pressure specific
    class hasPressure(DataProperty):
        """Pressure value"""
        domain = [PressureSensor]
        range = [float]
    
    # Current specific
    class hasCurrent(DataProperty):
        """Current value"""
        domain = [CurrentSensor]
        range = [float]
    
    # Equipment properties
    class hasSerialNumber(DataProperty):
        """Equipment serial number"""
        domain = [Equipment]
        range = [str]
    
    class hasModel(DataProperty):
        """Equipment model"""
        domain = [Equipment]
        range = [str]
    
    class hasManufacturer(DataProperty):
        """Equipment manufacturer"""
        domain = [Equipment]
        range = [str]
    
    # Failure properties
    class hasFailureCode(DataProperty):
        """Failure code"""
        domain = [Failure]
        range = [str]
    
    class hasSeverity(DataProperty):
        """Failure severity level"""
        domain = [Failure]
        range = [int]
    
    class hasDescription(DataProperty):
        """Description"""
        domain = [Thing]
        range = [str]
    
    # Anomaly properties
    class hasAnomalyScore(DataProperty):
        """Anomaly score"""
        domain = [Anomaly]
        range = [float]
    
    class hasThreshold(DataProperty):
        """Threshold value"""
        domain = [ThresholdAnomaly]
        range = [float]
    
    # Manufacturing specific properties
    class hasToolWear(DataProperty):
        """Tool wear measurement"""
        domain = [Tool]
        range = [float]
    
    class hasRotationalSpeed(DataProperty):
        """Rotational speed"""
        domain = [ManufacturingEquipment]
        range = [float]
    
    class hasTorque(DataProperty):
        """Torque measurement"""
        domain = [ManufacturingEquipment]
        range = [float]
    
    # Transportation specific properties
    class hasGPSLatitude(DataProperty):
        """GPS latitude"""
        domain = [PositionSensor]
        range = [float]
    
    class hasGPSLongitude(DataProperty):
        """GPS longitude"""
        domain = [PositionSensor]
        range = [float]
    
    class hasSpeed(DataProperty):
        """Speed measurement"""
        domain = [TransportationEquipment]
        range = [float]
    
    class hasFlowRate(DataProperty):
        """Flow rate measurement"""
        domain = [FlowSensor]
        range = [float]

print("âœ… Object properties and data properties defined")
print(f"ğŸ“Š Total object properties: {len(list(onto.object_properties()))}")
print(f"ğŸ“Š Total data properties: {len(list(onto.data_properties()))}")

print("\nğŸ”— OBJECT PROPERTIES:")
print("=" * 50)
for prop in onto.object_properties():
    print(f"ğŸ“‹ {prop.name}")

print("\nğŸ“Š DATA PROPERTIES:")
print("=" * 50)
for prop in onto.data_properties():
    print(f"ğŸ“‹ {prop.name}")
````

````python
# =================================================================
# TASK 2: NEO4J KNOWLEDGE GRAPH IMPLEMENTATION
# =================================================================

load_dotenv()
class KnowledgeGraphManager:
    """
    Knowledge Graph Manager for Neo4j implementation
    """
    
    def __init__(self, uri="bolt://localhost:7687", user="neo4j", password="Omkar@Neo4j123"):
        """
        Initialize Neo4j connection
        NOTE: Update connection details for your Neo4j instance
        """
        self.uri = uri
        self.user = user
        self.password = password
        self.driver = None
        
    def connect(self):
        """Connect to Neo4j database"""
        try:
            self.driver = GraphDatabase.driver(self.uri, auth=(self.user, self.password))
            print("âœ… Connected to Neo4j database")
            return True
        except Exception as e:
            print(f"âŒ Failed to connect to Neo4j: {e}")
            print("ğŸ’¡ Please ensure Neo4j is running and connection details are correct")
            return False
    
    def close(self):
        """Close Neo4j connection"""
        if self.driver:
            self.driver.close()
            print("âœ… Neo4j connection closed")
    
    def execute_query(self, query, parameters=None):
        """Execute a Cypher query"""
        if not self.driver:
            print("âŒ No database connection")
            return None
            
        with self.driver.session() as session:
            try:
                result = session.run(query, parameters or {})
                return result
            except Exception as e:
                print(f"âŒ Query execution failed: {e}")
                return None
    
    def clear_database(self):
        """Clear all nodes and relationships"""
        query = "MATCH (n) DETACH DELETE n"
        self.execute_query(query)
        print("ğŸ—‘ï¸ Database cleared")
    
    def create_constraints(self):
        """Create constraints for data integrity"""
        constraints = [
            # Equipment constraints
            "CREATE CONSTRAINT equipment_id IF NOT EXISTS FOR (e:Equipment) REQUIRE e.id IS UNIQUE",
            "CREATE CONSTRAINT manufacturing_equipment_id IF NOT EXISTS FOR (m:ManufacturingEquipment) REQUIRE m.id IS UNIQUE",
            "CREATE CONSTRAINT transportation_equipment_id IF NOT EXISTS FOR (t:TransportationEquipment) REQUIRE t.id IS UNIQUE",
            
            # Component constraints
            "CREATE CONSTRAINT component_id IF NOT EXISTS FOR (c:Component) REQUIRE c.id IS UNIQUE",
            "CREATE CONSTRAINT sensor_id IF NOT EXISTS FOR (s:Sensor) REQUIRE s.id IS UNIQUE",
            
            # Failure constraints
            "CREATE CONSTRAINT failure_id IF NOT EXISTS FOR (f:Failure) REQUIRE f.id IS UNIQUE",
            
            # Anomaly constraints
            "CREATE CONSTRAINT anomaly_id IF NOT EXISTS FOR (a:Anomaly) REQUIRE a.id IS UNIQUE"
        ]
        
        for constraint in constraints:
            try:
                self.execute_query(constraint)
                print(f"âœ… Constraint created: {constraint.split('FOR')[1].split('REQUIRE')[0].strip()}")
            except Exception as e:
                print(f"âš ï¸ Constraint already exists or failed: {e}")
    
    def create_indexes(self):
        """Create indexes for performance"""
        indexes = [
            "CREATE INDEX equipment_name IF NOT EXISTS FOR (e:Equipment) ON (e.name)",
            "CREATE INDEX sensor_type IF NOT EXISTS FOR (s:Sensor) ON (s.sensor_type)",
            "CREATE INDEX failure_type IF NOT EXISTS FOR (f:Failure) ON (f.failure_type)"
            # Removed generic timestamp index as it requires specific node label
        ]
        
        for index in indexes:
            try:
                self.execute_query(index)
                print(f"âœ… Index created: {index.split('ON')[1].strip() if 'ON' in index else 'General index'}")
            except Exception as e:
                print(f"âš ï¸ Index already exists or failed: {e}")

# Initialize Knowledge Graph Manager
kg_manager = KnowledgeGraphManager()

print("ğŸ”§ Neo4j Knowledge Graph Manager initialized")
print("ğŸ“ Connection details:")
print(f"   URI: {kg_manager.uri}")
print(f"   User: {kg_manager.user}")
print("\nğŸ’¡ To connect to Neo4j:")
print("   1. Ensure Neo4j Desktop/Server is running")
print("   2. Update connection details if needed")
print("   3. Run: kg_manager.connect()")

# Test connection (will fail if Neo4j is not running - that's expected)
print("\nğŸ” Testing Neo4j connection...")
if kg_manager.connect():
    print("ğŸ‰ Neo4j is ready!")
    
    # Create constraints and indexes
    print("\nğŸ“‹ Creating database constraints...")
    kg_manager.create_constraints()
    
    print("\nğŸ“Š Creating database indexes...")
    kg_manager.create_indexes()
    
    print("\nâœ… Neo4j database structure ready for population")
else:
    print("âš ï¸ Neo4j connection not available")
    print("   This is normal if Neo4j is not installed/running")
    print("   The knowledge graph code is ready to use when Neo4j becomes available")
````

````python
# =================================================================
# TASK 2.2: KNOWLEDGE GRAPH POPULATION FROM AI4I DATA
# =================================================================

class AI4IDataPopulator:
    """
    Populates knowledge graph with AI4I manufacturing data
    """
    
    def __init__(self, kg_manager):
        self.kg_manager = kg_manager
        self.data_dir = '/Users/omkarthorve/Desktop/poc_RCA/processed_data'
    
    def load_ai4i_data(self):
        """Load processed AI4I data"""
        try:
            ai4i_file = f"{self.data_dir}/ai4i_engineered.csv"
            self.ai4i_data = pd.read_csv(ai4i_file)
            print(f"âœ… Loaded AI4I data: {self.ai4i_data.shape}")
            return True
        except Exception as e:
            print(f"âŒ Failed to load AI4I data: {e}")
            return False
    
    def create_manufacturing_equipment(self):
        """Create manufacturing equipment nodes"""
        if not hasattr(self, 'ai4i_data'):
            print("âŒ AI4I data not loaded")
            return
        
        # Create unique equipment instances based on product types
        # Type was one-hot encoded, so we'll use the Type_H, Type_L, Type_M columns
        equipment_types = ['H', 'L', 'M']  # Product quality types
        
        for i, eq_type in enumerate(equipment_types):
            equipment_id = f"MFG_EQUIP_{eq_type}_{i+1:03d}"
            
            query = """
            CREATE (e:Equipment:ManufacturingEquipment {
                id: $equipment_id,
                name: $name,
                equipment_type: $eq_type,
                manufacturer: "Industrial Corp",
                model: $model,
                serial_number: $serial,
                created_at: datetime()
            })
            """
            
            parameters = {
                "equipment_id": equipment_id,
                "name": f"Manufacturing Line {eq_type}",
                "eq_type": eq_type,
                "model": f"ML-{eq_type}-2024",
                "serial": f"SN{i+1:06d}"
            }
            
            self.kg_manager.execute_query(query, parameters)
        
        print(f"âœ… Created {len(equipment_types)} manufacturing equipment nodes")
    
    def create_sensors(self):
        """Create sensor nodes"""
        sensor_definitions = [
            {"type": "TemperatureSensor", "name": "Air Temperature Sensor", "unit": "K"},
            {"type": "TemperatureSensor", "name": "Process Temperature Sensor", "unit": "K"},
            {"type": "CurrentSensor", "name": "Motor Current Sensor", "unit": "A"},
            {"type": "Sensor", "name": "Rotational Speed Sensor", "unit": "rpm"},
            {"type": "Sensor", "name": "Torque Sensor", "unit": "Nm"},
            {"type": "Sensor", "name": "Tool Wear Sensor", "unit": "min"}
        ]
        
        for i, sensor_def in enumerate(sensor_definitions):
            sensor_id = f"SENSOR_{i+1:03d}"
            
            query = f"""
            CREATE (s:Sensor:{sensor_def['type']} {{
                id: $sensor_id,
                name: $name,
                sensor_type: $sensor_type,
                unit: $unit,
                created_at: datetime()
            }})
            """
            
            parameters = {
                "sensor_id": sensor_id,
                "name": sensor_def["name"],
                "sensor_type": sensor_def["type"],
                "unit": sensor_def["unit"]
            }
            
            self.kg_manager.execute_query(query, parameters)
        
        print(f"âœ… Created {len(sensor_definitions)} sensor nodes")
    
    def create_failure_types(self):
        """Create failure type nodes"""
        failure_types = [
            {"type": "ToolWearFailure", "code": "TWF", "description": "Tool wear failure"},
            {"type": "HeatDissipationFailure", "code": "HDF", "description": "Heat dissipation failure"},
            {"type": "PowerFailure", "code": "PWF", "description": "Power failure"},
            {"type": "OverstrainFailure", "code": "OSF", "description": "Overstrain failure"},
            {"type": "RandomFailure", "code": "RNF", "description": "Random failure"}
        ]
        
        for failure_def in failure_types:
            failure_id = f"FAILURE_TYPE_{failure_def['code']}"
            
            query = f"""
            CREATE (f:Failure:{failure_def['type']} {{
                id: $failure_id,
                failure_type: $failure_type,
                failure_code: $code,
                description: $description,
                created_at: datetime()
            }})
            """
            
            parameters = {
                "failure_id": failure_id,
                "failure_type": failure_def["type"],
                "code": failure_def["code"],
                "description": failure_def["description"]
            }
            
            self.kg_manager.execute_query(query, parameters)
        
        print(f"âœ… Created {len(failure_types)} failure type nodes")
    
    def create_equipment_sensor_relationships(self):
        """Create relationships between equipment and sensors"""
        # Connect each manufacturing equipment to all sensors
        query = """
        MATCH (e:ManufacturingEquipment), (s:Sensor)
        CREATE (e)-[:HAS_SENSOR]->(s)
        CREATE (s)-[:MONITORS]->(e)
        """
        
        self.kg_manager.execute_query(query)
        print("âœ… Created equipment-sensor relationships")
    
    def populate_sample_data(self, sample_size=100):
        """Populate with sample sensor readings and failures"""
        if not hasattr(self, 'ai4i_data'):
            print("âŒ AI4I data not loaded")
            return
        
        # Reconstruct Type column from one-hot encoded columns
        def get_type(row):
            if row.get('Type_H', 0) == 1: return 'H'
            if row.get('Type_M', 0) == 1: return 'M'
            if row.get('Type_L', 0) == 1: return 'L'
            return 'M'  # default
        
        # Sample data for demonstration
        sample_data = self.ai4i_data.head(sample_size).copy()
        sample_data['Type'] = sample_data.apply(get_type, axis=1)
        
        for idx, row in sample_data.iterrows():
            # Create sensor reading nodes
            reading_id = f"READING_{idx:06d}"
            
            query = """
            CREATE (r:SensorReading {
                id: $reading_id,
                udi: $udi,
                air_temperature: $air_temp,
                process_temperature: $proc_temp,
                rotational_speed: $rot_speed,
                torque: $torque,
                tool_wear: $tool_wear,
                motor_current: $motor_current,
                timestamp: datetime(),
                product_type: $product_type
            })
            """
            
            parameters = {
                "reading_id": reading_id,
                "udi": int(row['UDI']),
                "air_temp": float(row['Air temperature [K]']),
                "proc_temp": float(row['Process temperature [K]']),
                "rot_speed": float(row['Rotational speed [rpm]']),
                "torque": float(row['Torque [Nm]']),
                "tool_wear": float(row['Tool wear [min]']),
                "motor_current": float(row.get('Motor_current', 0)),
                "product_type": str(row['Type'])
            }
            
            self.kg_manager.execute_query(query, parameters)
            
            # Create failure instances if machine failure occurred
            if row['Machine failure'] == 1:
                failure_id = f"FAILURE_INSTANCE_{idx:06d}"
                
                # Determine failure types
                failure_types = []
                if row['TWF'] == 1: failure_types.append('TWF')
                if row['HDF'] == 1: failure_types.append('HDF')
                if row['PWF'] == 1: failure_types.append('PWF')
                if row['OSF'] == 1: failure_types.append('OSF')
                if row['RNF'] == 1: failure_types.append('RNF')
                
                for ft in failure_types:
                    query = """
                    CREATE (fi:FailureInstance {
                        id: $failure_id,
                        udi: $udi,
                        failure_type: $failure_type,
                        severity: $severity,
                        timestamp: datetime()
                    })
                    """
                    
                    parameters = {
                        "failure_id": f"{failure_id}_{ft}",
                        "udi": int(row['UDI']),
                        "failure_type": ft,
                        "severity": 1
                    }
                    
                    self.kg_manager.execute_query(query, parameters)
        
        print(f"âœ… Populated {sample_size} sample records")

# Initialize AI4I populator
print("ğŸ­ Initializing AI4I Data Population...")
ai4i_populator = AI4IDataPopulator(kg_manager)

# Load and populate data (only if Neo4j is connected)
if kg_manager.driver:
    print("\nğŸ“ Loading AI4I data...")
    if ai4i_populator.load_ai4i_data():
        print("\nğŸ—ï¸ Creating manufacturing equipment nodes...")
        ai4i_populator.create_manufacturing_equipment()
        
        print("\nğŸ“¡ Creating sensor nodes...")
        ai4i_populator.create_sensors()
        
        print("\nâš ï¸ Creating failure type nodes...")
        ai4i_populator.create_failure_types()
        
        print("\nğŸ”— Creating relationships...")
        ai4i_populator.create_equipment_sensor_relationships()
        
        print("\nğŸ“Š Populating sample data...")
        ai4i_populator.populate_sample_data(100)
        
        print("\nâœ… AI4I data population completed!")
else:
    print("âš ï¸ Skipping data population - Neo4j not connected")
    print("   Run kg_manager.connect() first, then execute this cell")
````

````python
# =================================================================
# TASK 3: SWRL RULES REPOSITORY
# =================================================================

class SWRLRulesRepository:
    """
    Repository for SWRL rules covering manufacturing and transportation domains
    """
    
    def __init__(self):
        self.rules = {
            "manufacturing_rules": [],
            "transportation_rules": [],
            "cross_domain_rules": [],
            "physics_rules": [],
            "statistical_rules": [],
            "expert_rules": []
        }
        self.initialize_rules()
    
    def initialize_rules(self):
        """Initialize all rule categories"""
        self._create_manufacturing_rules()
        self._create_transportation_rules()
        self._create_physics_rules()
        self._create_statistical_rules()
        self._create_expert_rules()
        self._create_cross_domain_rules()
    
    def _create_manufacturing_rules(self):
        """Manufacturing domain specific rules (AI4I)"""
        
        # Tool Wear Failure Rules
        self.rules["manufacturing_rules"].extend([
            {
                "id": "MFG_R001",
                "name": "Tool Wear Threshold Rule",
                "description": "If tool wear exceeds threshold, predict tool wear failure",
                "swrl": "ManufacturingEquipment(?e) âˆ§ hasToolWear(?e, ?wear) âˆ§ swrlb:greaterThan(?wear, 200) â†’ hasFailure(?e, ToolWearFailure)",
                "cypher_equivalent": """
                MATCH (e:ManufacturingEquipment)
                WHERE e.tool_wear > 200
                CREATE (e)-[:HAS_FAILURE]->(f:Failure:ToolWearFailure {
                    id: 'TWF_' + e.id + '_' + toString(timestamp()),
                    predicted_by: 'Tool_Wear_Threshold_Rule',
                    confidence: 0.85
                })
                """,
                "domain": "manufacturing",
                "category": "threshold"
            },
            {
                "id": "MFG_R002", 
                "name": "Heat Dissipation Failure Rule",
                "description": "If process temperature significantly higher than air temperature, risk of heat dissipation failure",
                "swrl": "ManufacturingEquipment(?e) âˆ§ hasProcessTemperature(?e, ?pt) âˆ§ hasAirTemperature(?e, ?at) âˆ§ swrlb:subtract(?diff, ?pt, ?at) âˆ§ swrlb:greaterThan(?diff, 10) â†’ hasFailure(?e, HeatDissipationFailure)",
                "cypher_equivalent": """
                MATCH (e:ManufacturingEquipment)
                WHERE e.process_temperature - e.air_temperature > 10
                CREATE (e)-[:HAS_FAILURE]->(f:Failure:HeatDissipationFailure {
                    id: 'HDF_' + e.id + '_' + toString(timestamp()),
                    predicted_by: 'Heat_Dissipation_Rule',
                    confidence: 0.78
                })
                """,
                "domain": "manufacturing",
                "category": "physics"
            },
            {
                "id": "MFG_R003",
                "name": "Power Failure Rule", 
                "description": "If torque is low but rotational speed is high, indicates power issues",
                "swrl": "ManufacturingEquipment(?e) âˆ§ hasTorque(?e, ?torque) âˆ§ hasRotationalSpeed(?e, ?speed) âˆ§ swrlb:lessThan(?torque, 30) âˆ§ swrlb:greaterThan(?speed, 1600) â†’ hasFailure(?e, PowerFailure)",
                "cypher_equivalent": """
                MATCH (e:ManufacturingEquipment)  
                WHERE e.torque < 30 AND e.rotational_speed > 1600
                CREATE (e)-[:HAS_FAILURE]->(f:Failure:PowerFailure {
                    id: 'PWF_' + e.id + '_' + toString(timestamp()),
                    predicted_by: 'Power_Failure_Rule',
                    confidence: 0.72
                })
                """,
                "domain": "manufacturing",
                "category": "physics"
            },
            {
                "id": "MFG_R004",
                "name": "Quality-Based Maintenance Rule",
                "description": "Low quality products require more frequent maintenance",
                "swrl": "ManufacturingEquipment(?e) âˆ§ hasQualityType(?e, ?qt) âˆ§ swrlb:equal(?qt, 'L') â†’ requiresMaintenance(?e, PreventiveMaintenance)",
                "cypher_equivalent": """
                MATCH (e:ManufacturingEquipment)
                WHERE e.product_type = 'L'
                CREATE (e)-[:REQUIRES_MAINTENANCE]->(m:MaintenanceAction:PreventiveMaintenance {
                    id: 'PREV_MAINT_' + e.id + '_' + toString(timestamp()),
                    reason: 'Low_Quality_Production',
                    priority: 'high'
                })
                """,
                "domain": "manufacturing", 
                "category": "expert"
            }
        ])
    
    def _create_transportation_rules(self):
        """Transportation domain specific rules (MetroPT)"""
        
        self.rules["transportation_rules"].extend([
            {
                "id": "TRANS_R001",
                "name": "Compressor Failure Rule",
                "description": "If compressor status is off and pressure is low, compressor failure likely",
                "swrl": "TransportationEquipment(?e) âˆ§ hasComponent(?e, ?comp) âˆ§ Compressor(?comp) âˆ§ hasStatus(?comp, 0) âˆ§ hasPressure(?e, ?pressure) âˆ§ swrlb:lessThan(?pressure, -0.05) â†’ hasFailure(?comp, CompressorFailure)",
                "cypher_equivalent": """
                MATCH (e:TransportationEquipment)-[:HAS_COMPONENT]->(c:Compressor)
                WHERE c.status = 0 AND e.pressure < -0.05
                CREATE (c)-[:HAS_FAILURE]->(f:Failure {
                    id: 'COMP_FAIL_' + c.id + '_' + toString(timestamp()),
                    failure_type: 'CompressorFailure',
                    predicted_by: 'Compressor_Failure_Rule',
                    confidence: 0.88
                })
                """,
                "domain": "transportation",
                "category": "expert"
            },
            {
                "id": "TRANS_R002",
                "name": "Oil Temperature Anomaly Rule",
                "description": "If oil temperature exceeds normal operating range, create anomaly",
                "swrl": "TransportationEquipment(?e) âˆ§ hasOilTemperature(?e, ?temp) âˆ§ swrlb:greaterThan(?temp, 70) â†’ hasAnomaly(?e, TemperatureAnomaly)",
                "cypher_equivalent": """
                MATCH (e:TransportationEquipment)
                WHERE e.oil_temperature > 70
                CREATE (e)-[:HAS_ANOMALY]->(a:Anomaly:TemperatureAnomaly {
                    id: 'TEMP_ANOM_' + e.id + '_' + toString(timestamp()),
                    anomaly_type: 'TemperatureAnomaly',
                    value: e.oil_temperature,
                    threshold: 70,
                    severity: 'high'
                })
                """,
                "domain": "transportation",
                "category": "threshold"
            },
            {
                "id": "TRANS_R003",
                "name": "Motor Current Spike Rule",
                "description": "Sudden increase in motor current indicates potential motor issues",
                "swrl": "TransportationEquipment(?e) âˆ§ hasMotorCurrent(?e, ?current) âˆ§ swrlb:greaterThan(?current, 5.0) â†’ hasAnomaly(?e, CurrentAnomaly)",
                "cypher_equivalent": """
                MATCH (e:TransportationEquipment)
                WHERE e.motor_current > 5.0
                CREATE (e)-[:HAS_ANOMALY]->(a:Anomaly:CurrentAnomaly {
                    id: 'CURR_ANOM_' + e.id + '_' + toString(timestamp()),
                    anomaly_type: 'CurrentAnomaly', 
                    value: e.motor_current,
                    threshold: 5.0,
                    severity: 'medium'
                })
                """,
                "domain": "transportation",
                "category": "threshold"
            }
        ])
    
    def _create_physics_rules(self):
        """Physics-based rules applicable across domains"""
        
        self.rules["physics_rules"].extend([
            {
                "id": "PHYS_R001",
                "name": "Thermal Expansion Rule",
                "description": "High temperature differences cause thermal stress",
                "swrl": "Equipment(?e) âˆ§ hasTemperatureDifferential(?e, ?diff) âˆ§ swrlb:greaterThan(?diff, 15) â†’ hasAnomaly(?e, ThermalStressAnomaly)",
                "cypher_equivalent": """
                MATCH (e:Equipment)
                WHERE e.temperature_differential > 15
                CREATE (e)-[:HAS_ANOMALY]->(a:Anomaly {
                    id: 'THERMAL_STRESS_' + e.id + '_' + toString(timestamp()),
                    anomaly_type: 'ThermalStressAnomaly',
                    cause: 'High_Temperature_Differential'
                })
                """,
                "domain": "cross",
                "category": "physics"
            },
            {
                "id": "PHYS_R002",
                "name": "Energy Conservation Rule",
                "description": "Power consumption should correlate with work output",
                "swrl": "Equipment(?e) âˆ§ hasPowerConsumption(?e, ?power) âˆ§ hasWorkOutput(?e, ?work) âˆ§ hasEfficiency(?e, ?eff) âˆ§ swrlb:divide(?calc_eff, ?work, ?power) âˆ§ swrlb:lessThan(?calc_eff, 0.5) â†’ hasAnomaly(?e, EfficiencyAnomaly)",
                "cypher_equivalent": """
                MATCH (e:Equipment)
                WHERE e.work_output / e.power_consumption < 0.5
                CREATE (e)-[:HAS_ANOMALY]->(a:Anomaly {
                    id: 'EFF_ANOM_' + e.id + '_' + toString(timestamp()),
                    anomaly_type: 'EfficiencyAnomaly',
                    efficiency: e.work_output / e.power_consumption
                })
                """,
                "domain": "cross",
                "category": "physics"
            }
        ])
    
    def _create_statistical_rules(self):
        """Statistical rules for anomaly detection"""
        
        self.rules["statistical_rules"].extend([
            {
                "id": "STAT_R001",
                "name": "Moving Average Deviation Rule",
                "description": "If current value deviates significantly from moving average, create anomaly",
                "swrl": "Sensor(?s) âˆ§ hasValue(?s, ?value) âˆ§ hasMovingAverage(?s, ?avg) âˆ§ hasStandardDeviation(?s, ?std) âˆ§ swrlb:subtract(?diff, ?value, ?avg) âˆ§ swrlb:abs(?abs_diff, ?diff) âˆ§ swrlb:multiply(?threshold, ?std, 2) âˆ§ swrlb:greaterThan(?abs_diff, ?threshold) â†’ hasAnomaly(?s, StatisticalAnomaly)",
                "cypher_equivalent": """
                MATCH (s:Sensor)
                WHERE abs(s.current_value - s.moving_average) > 2 * s.standard_deviation
                CREATE (s)-[:HAS_ANOMALY]->(a:Anomaly:StatisticalAnomaly {
                    id: 'STAT_ANOM_' + s.id + '_' + toString(timestamp()),
                    z_score: abs(s.current_value - s.moving_average) / s.standard_deviation,
                    threshold: 2.0
                })
                """,
                "domain": "cross",
                "category": "statistical"
            }
        ])
    
    def _create_expert_rules(self):
        """Expert knowledge rules from domain specialists"""
        
        self.rules["expert_rules"].extend([
            {
                "id": "EXP_R001",
                "name": "Maintenance Schedule Rule",
                "description": "Equipment with multiple anomalies requires immediate maintenance",
                "swrl": "Equipment(?e) âˆ§ hasAnomaly(?e, ?a1) âˆ§ hasAnomaly(?e, ?a2) âˆ§ swrlb:notEqual(?a1, ?a2) â†’ requiresMaintenance(?e, ImmediateMaintenance)",
                "cypher_equivalent": """
                MATCH (e:Equipment)
                WHERE size((e)-[:HAS_ANOMALY]->()) >= 2
                CREATE (e)-[:REQUIRES_MAINTENANCE]->(m:MaintenanceAction {
                    id: 'IMM_MAINT_' + e.id + '_' + toString(timestamp()),
                    maintenance_type: 'ImmediateMaintenance',
                    priority: 'critical',
                    reason: 'Multiple_Anomalies'
                })
                """,
                "domain": "cross",
                "category": "expert"
            }
        ])
    
    def _create_cross_domain_rules(self):
        """Rules that apply across both domains"""
        
        self.rules["cross_domain_rules"].extend([
            {
                "id": "CROSS_R001", 
                "name": "Cascade Failure Prevention Rule",
                "description": "If one component fails, check related components for anomalies",
                "swrl": "Equipment(?e) âˆ§ hasComponent(?e, ?c1) âˆ§ hasComponent(?e, ?c2) âˆ§ hasFailure(?c1, ?f) âˆ§ swrlb:notEqual(?c1, ?c2) â†’ requiresInspection(?c2)",
                "cypher_equivalent": """
                MATCH (e:Equipment)-[:HAS_COMPONENT]->(c1:Component)-[:HAS_FAILURE]->()
                MATCH (e)-[:HAS_COMPONENT]->(c2:Component)
                WHERE c1 <> c2
                CREATE (c2)-[:REQUIRES_INSPECTION]->(i:Inspection {
                    id: 'INSP_' + c2.id + '_' + toString(timestamp()),
                    reason: 'Cascade_Failure_Prevention',
                    priority: 'high'
                })
                """,
                "domain": "cross",
                "category": "expert"
            }
        ])
    
    def get_rules_by_domain(self, domain):
        """Get rules for specific domain"""
        if domain in self.rules:
            return self.rules[domain]
        return []
    
    def get_rules_by_category(self, category):
        """Get rules by category"""
        result = []
        for domain_rules in self.rules.values():
            result.extend([rule for rule in domain_rules if rule.get('category') == category])
        return result
    
    def export_rules(self, output_dir):
        """Export rules to files"""
        rules_file = f"{output_dir}/swrl_rules.json"
        with open(rules_file, 'w') as f:
            json.dump(self.rules, f, indent=2)
        
        # Create SWRL-only file
        swrl_file = f"{output_dir}/rules.swrl"
        with open(swrl_file, 'w') as f:
            f.write("// SWRL Rules for Predictive Maintenance\\n\\n")
            for domain, domain_rules in self.rules.items():
                f.write(f"// {domain.upper()} RULES\\n")
                for rule in domain_rules:
                    f.write(f"// {rule['id']}: {rule['name']}\\n")
                    f.write(f"{rule['swrl']}\\n\\n")
        
        # Create Cypher queries file
        cypher_file = f"{output_dir}/neo4j_rules.cypher"
        with open(cypher_file, 'w') as f:
            f.write("// Neo4j Cypher Rule Implementations\\n\\n")
            for domain, domain_rules in self.rules.items():
                f.write(f"// {domain.upper()} RULES\\n")
                for rule in domain_rules:
                    f.write(f"// {rule['id']}: {rule['name']}\\n")
                    f.write(f"{rule['cypher_equivalent']}\\n\\n")
        
        return rules_file, swrl_file, cypher_file

# Initialize SWRL Rules Repository
print("ğŸ“š Initializing SWRL Rules Repository...")
rules_repo = SWRLRulesRepository()

# Export rules
rules_file, swrl_file, cypher_file = rules_repo.export_rules(rules_dir)

print("âœ… SWRL Rules Repository created")
print(f"ğŸ“Š Total rules by domain:")
for domain, domain_rules in rules_repo.rules.items():
    print(f"   - {domain}: {len(domain_rules)} rules")

print(f"\\nğŸ“ Rules exported to:")
print(f"   - JSON: {rules_file}")
print(f"   - SWRL: {swrl_file}")
print(f"   - Cypher: {cypher_file}")

print("\\nğŸ” Sample Manufacturing Rule:")
sample_rule = rules_repo.rules["manufacturing_rules"][0]
print(f"   ID: {sample_rule['id']}")
print(f"   Name: {sample_rule['name']}")
print(f"   SWRL: {sample_rule['swrl']}")
````
