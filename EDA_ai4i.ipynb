```python
# Convert AI4I to pseudo-time series using sliding windows
def create_time_series_windows(df, window_size=10, step_size=1):
    """
    Convert static AI4I dataset to pseudo-time series using sliding windows
    """
    # Sort by UDI to create artificial temporal order
    df_sorted = df.sort_values('UDI').reset_index(drop=True)
    
    windowed_data = []
    for i in range(0, len(df_sorted) - window_size + 1, step_size):
        window = df_sorted.iloc[i:i+window_size].copy()
        # Add temporal features
        window['time_step'] = range(window_size)
        window['window_id'] = i // step_size
        windowed_data.append(window)
    
    return pd.concat(windowed_data, ignore_index=True)

# Create windowed dataset
WINDOW_SIZE = 10
STEP_SIZE = 5
ai4i_windowed = create_time_series_windows(ai4i_df, WINDOW_SIZE, STEP_SIZE)
print(f"Original dataset: {ai4i_df.shape}")
print(f"Windowed dataset: {ai4i_windowed.shape}")
```

```python
# Unified Feature Engineering Pipeline
def feature_engineering_pipeline(df):
    """
    Standardized feature engineering for AI4I dataset
    """
    df_processed = df.copy()
    
    # 1. Normalization (Z-score)
    numerical_features = ['Air temperature [K]', 'Process temperature [K]', 
                         'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
    
    for feature in numerical_features:
        df_processed[f'{feature}_normalized'] = (df_processed[feature] - df_processed[feature].mean()) / df_processed[feature].std()
    
    # 2. Feature Creation
    df_processed['temp_difference'] = df_processed['Process temperature [K]'] - df_processed['Air temperature [K]']
    df_processed['power_estimate'] = df_processed['Torque [Nm]'] * df_processed['Rotational speed [rpm]'] / 9549  # Power in kW
    df_processed['thermal_stress'] = df_processed['Process temperature [K]'] / df_processed['Air temperature [K]']
    
    # 3. Categorical encoding
    df_processed = pd.get_dummies(df_processed, columns=['Type'], prefix='Type')
    
    return df_processed

# Apply feature engineering
ai4i_engineered = feature_engineering_pipeline(ai4i_df)
print(f"Features after engineering: {ai4i_engineered.shape[1]} columns")
```

```python
# Data Quality and Compatibility Assessment
def generate_data_quality_report(df, df_windowed, df_engineered):
    """
    Generate comprehensive data quality report
    """
    report = {
        'original_dataset': {
            'shape': df.shape,
            'missing_values': df.isnull().sum().sum(),
            'data_types': df.dtypes.value_counts().to_dict(),
            'memory_usage_mb': df.memory_usage(deep=True).sum() / 1024**2
        },
        'windowed_dataset': {
            'shape': df_windowed.shape,
            'compression_ratio': df_windowed.shape[0] / df.shape[0],
            'temporal_features_added': ['time_step', 'window_id']
        },
        'engineered_dataset': {
            'shape': df_engineered.shape,
            'new_features_count': df_engineered.shape[1] - df.shape[1],
            'feature_categories': {
                'normalized': [col for col in df_engineered.columns if '_normalized' in col],
                'derived': ['temp_difference', 'power_estimate', 'thermal_stress'],
                'encoded': [col for col in df_engineered.columns if 'Type_' in col]
            }
        }
    }
    return report

# Generate report
quality_report = generate_data_quality_report(ai4i_df, ai4i_windowed, ai4i_engineered)
print("="*60)
print("DATA QUALITY & COMPATIBILITY REPORT")
print("="*60)
for key, value in quality_report.items():
    print(f"\n{key.upper()}:")
    for subkey, subvalue in value.items():
        print(f"  {subkey}: {subvalue}")
```

```python
# Save processed datasets and documentation
import os

# Create output directory
output_dir = '/Users/omkarthorve/Desktop/poc_RCA/processed_data'
os.makedirs(output_dir, exist_ok=True)

# Save datasets
ai4i_windowed.to_csv(f'{output_dir}/ai4i_windowed.csv', index=False)
ai4i_engineered.to_csv(f'{output_dir}/ai4i_engineered.csv', index=False)

# Save quality report (convert pandas dtypes to serializable format)
import json

# Convert quality report to JSON-serializable format
def make_json_serializable(obj):
    if hasattr(obj, 'to_dict'):
        return obj.to_dict()
    elif hasattr(obj, '__dict__'):
        return str(obj)
    else:
        return str(obj)

# Create a clean copy of the report for JSON serialization
json_quality_report = {}
for key, value in quality_report.items():
    json_quality_report[key] = {}
    for subkey, subvalue in value.items():
        if subkey == 'data_types':
            # Convert dtype counts to string keys
            json_quality_report[key][subkey] = {str(k): v for k, v in subvalue.items()}
        else:
            json_quality_report[key][subkey] = subvalue

with open(f'{output_dir}/data_quality_report.json', 'w') as f:
    json.dump(json_quality_report, f, indent=2, default=str)

print("Phase 1 Deliverables saved:")
print("✅ Clean, windowed AI4I dataset")
print("✅ Unified data preprocessing pipeline")
print("✅ Data quality and compatibility report")
```
